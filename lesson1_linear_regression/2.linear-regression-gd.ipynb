# 导入相关包

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline
```

# 初始化数据和权重

```python
x = np.linspace(-10, 10, 100)
w_true = 3.0
b_true = 4.0
#y_true = w_true * x + b_true + np.random.normal(size=(len(x), )) * 3
y_true = w_true * x + b_true
plt.scatter(x, y_true)
```

# 定义目标函数

```python
def objective(w, b, x, y):
    y_pred = w * x + b
    #print y_pred.shape
    #print y_true.shape
    loss = np.mean((y - y_pred) ** 2)
    return loss
print objective(w_true, b_true, x, y_true)
```

# 目标函数关于w,b的图像

```python
def plot_surface(w, b, x, y):
    all_loss = np.zeros((len(w), len(b)))
    for i in range(len(w)):
        for j in range(len(b)):
            loss = objective(w[i], b[i], x, y)
            all_loss[i][j] = loss
    w_grid, b_grid = np.meshgrid(w, b)
    return w_grid, b_grid, all_loss
```

```python
w_list = np.linspace(-20, 20, 100)
b_list = np.linspace(-20, 20, 100)
w_grid, b_grid, loss_grid = plot_surface(w_list, b_list, x, y_true)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_top_view()
ax.plot_surface(w_grid, b_grid, loss_grid, rstride=2, cstride=2)
```

# 定义梯度函数、模型训练

```python
def dw(w, b, x, y):
    return np.mean((w * x + b - y) * x) * 2

def db(w, b, x, y):
    return np.mean(w * x + b - y) * 2

w = -1
b = 0
print "total loss for init w b:", objective(w, b, x, y_true)

# 超参数
epochs = 30
lr = 0.01
batch_size = 10
decay_factor = 1
lr_decay = 2

losses = []
for i in xrange(epochs):
    epoch_loss = 0
    if lr_decay and i > lr_decay:
        lr = lr * decay_factor
    idx = range(len(x))
    random.shuffle(idx)
    for index in xrange(0, len(idx), batch_size):
        j = np.array(idx[index:min(index+batch_size, len(idx))])
        x_batch = x.take(j)
        y_batch = y_true.take(j)
        #print x_batch
        #print y_batch
        batch_loss = objective(w, b, x_batch, y_batch)
        #print "batch %d loss is" % index, batch_loss
        epoch_loss += batch_loss
        w = w - lr * dw(w, b, x_batch, y_batch)
        b = b - lr * db(w, b, x_batch, y_batch)
        #print "w is:",w
        #print "b is:",b
    losses.append(epoch_loss)
plt.plot(losses)

print "minimum epoch loss is:", min(losses)
print "true w b is:", w_true, b_true
print "predicted w b is:", w, b
```
